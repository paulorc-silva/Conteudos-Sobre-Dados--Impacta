````python
distorcao = []

for i in range(1, 11):
    km = KMeans(n_clusters=i)
    km.fit(x_resampled)
    distorcao.append(km.inertia_)

plt.plot(range(1, 11), distorcao, marker='o')
plt.xlabel('N clusters')
plt.ylabel('Distorção')
plt.show()
````

````python
warnings.filterwarnings('ignore')
for n_clusters in range(2, 11):
    clusterer = KMeans(n_clusters=n_clusters,)
    cluster_labels = clusterer.fit_predict(x_resampled)

    silhouette_avg = silhouette_score(x_resampled, cluster_labels)
    print(f'Para {n_clusters} clusters - Score de silhueta: {silhouette_avg}')
````
